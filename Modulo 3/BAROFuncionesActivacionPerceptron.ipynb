# Definiciones de Funciones

```python
# Importar la libreria de las funciones matematicas
import numpy as np
```

```python
# Esta es la definicnion de la función escalon unitario
# h será la suma ponderada o combinación lineal

def escalon_unitario(x):
    if x >= 0:
        return 1
    else:
        return 0
```

```python
# Se define la función sigmoide
def sigmoide(x):
    sig = 1/(1+np.exp(-x))
    return sig
```

```python
# Esta es la definición tangente hiperbolica
# numpy tiene un método que calcula esta función 
def tangenteh(x):
    tan = np.tanh(x)
    return tan
```

```python
# Definición de la función reLU
def relu(x):
    return max(0,x)
```

# Compuerta AND

```python
# definición de tabla de verdad (2x4)
inputs = np.array([[0,0],[0,1],[1,0],[1,1]])
targets = np.array([0,0,0,1])
```

```python
weights = np.array([1,1])
bias = -1.5
```

```python
outputs = []
for i in range(targets.shape[0]):
    h = np.dot(inputs[i], weights) + bias #suma ponderada
    y = escalon_unitario(h)
    outputs.append([h,y])

print('Escalon Unitario')
print ('input 1','input 2','Combinación lineal', 'Target', 'Output')
for i in range(4):
    print('  ',inputs[i][0],'    ', inputs[i][1], '        ', (round(outputs[i][0],2)),'            ', targets[i], '    ', outputs[i][1])

outputs = []
for i in range(targets.shape[0]):
    h = np.dot(inputs[i], weights) + bias #suma ponderada
    y = sigmoide(h)
    outputs.append([h,y])

print('Sigmoide')
print ('input 1','input 2','Combinación lineal', 'Target', 'Output')
for i in range(4):
    print('  ',inputs[i][0],'    ', inputs[i][1], '        ', (round(outputs[i][0],2)),'            ', targets[i], '    ', (round(outputs[i][1],2)))

outputs = []
for i in range(targets.shape[0]):
    h = np.dot(inputs[i], weights) + bias #suma ponderada
    y = relu(h)
    outputs.append([h,y])

print('ReLU')
print ('input 1','input 2','Combinación lineal', 'Target', 'Output')
for i in range(4):
    print('  ',inputs[i][0],'    ', inputs[i][1], '        ', (round(outputs[i][0],2)),'            ', targets[i], '    ', outputs[i][1])
```
